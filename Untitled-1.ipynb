```python
import pandas as pd

# Create the dataframes
df1 = pd.DataFrame({
    'data1': ['a1', 'a2'],
    'data2': ['b1', 'b2'],
    'data3': ['c1', 'c2']
})

df2 = pd.DataFrame({
    'data1': ['a1', 'a2', 'a2'],
    'data2': ['b1', 'b2', 'b3']
})
print(df2.columns)
print(df1)
print(df2)
# Filter rows in df2 where [data1, data2] do not exist in df1
filtered_df = df2[~df2[['data1', 'data2']].apply(tuple, 1).isin(df1[['data1', 'data2']].apply(tuple, 1))]

# Display the final dataframe
print(filtered_df)

```

```python
import altair as alt
import pandas as pd

# Sample DataFrame with decimal values
df = pd.DataFrame({
    'category': ['A', 'B', 'C'],
    'value': [1.5, 2.3, 3.8]  # Decimal values
})

# Create a chart and specify the correct data type for the 'value' column
chart = alt.Chart(df).mark_bar().encode(
    x='category:O',  # Categorical data, so 'O' for ordinal or nominal
    y=alt.Y('value:Q')  # Quantitative data, so 'Q' for numeric types
)
chart.show()
```
